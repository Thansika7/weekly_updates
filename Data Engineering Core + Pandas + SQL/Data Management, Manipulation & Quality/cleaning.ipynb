```python
import pandas as pd
df = pd.read_csv("audible_uncleaned.csv")
```

```python
df.info()
df.head()
```

```python
df.duplicated().any()
```

```python
df['author'] = df['author'].str.replace('Writtenby:', '')
df['narrator'] = df['narrator'].str.replace('Narratedby:', '')
```

```python
df.head()
```

```python
df['price'] = df['price'].str.replace(',', '')
df['price'] = df['price'].replace('Free', '0')
df['price'] = df['price'].astype(float)
```

```python
df[df['price'] % 1 != 0]
```

```python
df.head()
```

```python
df.head()
```

```python
df['time'] = df['time'].replace('Less than 1 minute', '1 min')
df['hours'] = df['time'].str.extract(r'(\d+)\s*hr').fillna(0).astype(int)
df['minutes'] = df['time'].str.extract(r'(\d+)\s*min').fillna(0).astype(int)
df['duration_minutes'] = df['hours'] * 60 + df['minutes']
print(df[['time', 'duration_minutes']].head())
```

```python
df.head()
```

```python

df.drop(columns=['hours', 'minutes'], inplace=True)
df.drop(columns=['time'], inplace=True)
```

```python
df.head()
```

```python
df['author'] = df['author'].str.replace(r'([a-z])([A-Z])', r'\1 \2', regex=True)
df['narrator'] = df['narrator'].str.replace(r'([a-z])([A-Z])', r'\1 \2', regex=True)

```

```python
df.columns = df.columns.str.upper()
```

```python
df.isna().sum()
```

```python
df['STAR_VALUE'] = df['STARS'].str.extract(r'([\d.]+)').astype(float)
```

```python
df['RATING_PERCENT'] = (df['STAR_VALUE'] / 5) * 100
```

```python
df[['STAR_VALUE','RATING_PERCENT']] = df[['STAR_VALUE','RATING_PERCENT']].fillna(0).astype(int)
```

```python
df.drop(columns=['STARS','STAR_VALUE'], inplace=True)
```

```python
df = df.sort_values(by='RATING_PERCENT', ascending=False)
```

```python
df.to_csv("audible_cleaned.csv", index=False)
```
